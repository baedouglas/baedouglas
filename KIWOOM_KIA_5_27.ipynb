# 기아 타순 정하기

```python
Clutch_Ability_KIA = Position_KIA[["1루,2루", "1루,3루", "2루,3루", "만루"]].mean(axis=1)

Count_Features_KIA = pd.DataFrame(index=range(10))
Count_Features_KIA["초기카운트"] = Count_KIA[["0S 0B", "1S ", "1S 1B"]].mean(axis=1)
Count_Features_KIA["스트라이크카운트"] = Count_KIA[["2S", "1S 2B", "2S 2B", "2S 3B"]].mean(axis=1)
Count_Features_KIA["풀카운트"] = Count_KIA[["2S 3B", "3B"]].mean(axis=1)

Final_Features_KIA = pd.DataFrame({
    "OPS": Comprehensive_KIA["OPS"],
    "출루율": Comprehensive_KIA["출루율"],
    "장타율": Comprehensive_KIA["장타율"],
    "0OUT": Outcount_KIA["0 OUT"],
    "2OUT": Outcount_KIA["2 OUT"],
    "득점권": Clutch_Ability_KIA,
    "볼넷": Comprehensive_KIA["볼넷"] / Comprehensive_KIA["볼넷"].max(),
    "삼진": Comprehensive_KIA["삼진"] / Comprehensive_KIA["삼진"].max(),
    "도루": Comprehensive_KIA["도루"] / Comprehensive_KIA["도루"].max(),
    "초기카운트": Count_Features_KIA["초기카운트"],
    "스트라이크카운트": Count_Features_KIA["스트라이크카운트"],
    "풀카운트": Count_Features_KIA["풀카운트"],
    "투수별": Pitcher_KIA["하영민"],
    "포수특수" : Comprehensive_KIA["포수특수"]
})

Final_Features_KIA
```

### Weights = [OPS, 출루율, 장타율, 0OUT, 2OUT, 득점권, 볼넷, 삼진, 도루, 초기카운트, 스트라이크 카운트, 풀카운트,투수별, 포수특수]
1번타자 : 출루 잘 해야함, 빠름 -> 출루율 높음, 삼진 적어야함, 도루율 높음  
2번타자 : 1번타자와 연결, 빠름 -> 출루율 높음, 삼진 적어야함, 도루율 높음  
3번타자 : 득점권에 주자가 있을때 타격 성과 고려 -> OPS, 장타율  
4번타자 : 해결사 -> OPS, 장타율, 득점권  
5번타자 : 4번타자에 이은 후속타 -> OPS, 득점권  
6번타자 : 2번타자와 마찬가지로 안정적 연결 -> 출루율, 스탯 골고루 중요  |
7번타자 : 기동력 + 잔류 최소화 -> 스탯 골고루 중요  
8번타자 : 희생플레이 -> 특별한 스택 중요도 없음  
9번타자 : 1번타자와 이어지므로 출루율 중요  

```python
Weights = {
    "1번": [0.10, 0.30, 0.05, 0.20, 0.05, 0.00, 0.15, -0.20, 0.30, 0.10, 0.15, 0.10, 0.15, 0.00],  
    "2번": [0.15, 0.25, 0.10, 0.10, 0.10, 0.05, 0.10, -0.10, 0.30, 0.10, 0.10, 0.05, 0.15, 0.00],  
    "3번": [0.30, 0.20, 0.20, 0.05, 0.10, 0.10, 0.05, -0.05, 0.05, 0.05, 0.05, 0.05, 0.15, 0.00], 
    "4번": [0.35, 0.05, 0.30, 0.00, 0.05, 0.20, 0.00, 0.00, 0.05, 0.00, 0.00, 0.05, 0.15, 0.00],  
    "5번": [0.25, 0.10, 0.20, 0.05, 0.10, 0.20, 0.00, 0.00, 0.05, 0.05, 0.05, 0.05, 0.15, 0.00], 
    "6번": [0.15, 0.15, 0.10, 0.10, 0.15, 0.10, 0.10, -0.10, 0.05, 0.10, 0.05, 0.00, 0.15, 0.00], 
    "7번": [0.10, 0.10, 0.05, 0.10, 0.10, 0.05, 0.10, -0.05, 0.05, 0.10, 0.05, 0.05, 0.15, 0.00], 
    "8번": [0.05, 0.10, 0.05, 0.10, 0.10, 0.10, 0.10, -0.05, 0.05, 0.10, 0.10, 0.05, 0.15, 0.90], 
    "9번": [0.10, 0.25, 0.05, 0.20, 0.10, 0.10, 0.10, -0.05, 0.05, 0.10, 0.10, 0.05, 0.15, 0.00], 
}

# 점수 계산
Score_Table_KIA = pd.DataFrame(index=Final_Features_KIA.index)
for role, w in Weights.items():
    Score_Table_KIA[role] = Final_Features_KIA.values @ np.array(w)

# 타자 중복 없이 배치
Final_Lineup_KIA = {}
used_KIA = set()
for role in Score_Table_KIA.columns:
    scores = Score_Table_KIA[role].copy()
    scores.loc[list(used_KIA)] = -float("inf")
    pick = scores.idxmax()
    Final_Lineup_KIA[role] = pick
    used_KIA.add(pick)

# 결과 정리
Final_Lineup_KIA_df = pd.DataFrame(list(Final_Lineup_KIA.items()), columns=["타순", "타자 인덱스"]).set_index("타순")
Final_Lineup_KIA_df
```

```python
def compute_player_hit_probs(df):

    # 타율과 OPS의 최소, 최대 값 계산합니다.
    avg_min, avg_max = df['타율'].min(), df['타율'].max()
    ops_min, ops_max = df['OPS'].min(), df['OPS'].max()

    player_probs = {}

    for idx, row in df.iterrows():
        pa = estimate_at_bats(row)
        total_hits = row['안타']

        if pa == 0 or total_hits == 0 or pd.isna(row['타율']) or pd.isna(row['OPS']):
            probs = {'1B': 0, '2B': 0, '3B': 0, 'HR': 0, 'OUT': 1.0}
            
        else:
            avg_range = max(0.001, avg_max - avg_min)
            ops_range = max(0.001, ops_max - ops_min)

            avg_norm = (row['타율'] - avg_min) / avg_range
            ops_norm = (row['OPS'] - ops_min) / ops_range

            # 선수별 안타칠 확률을 계산합니다. (0.1~0.3 사이로 설정)
            # 이 값들을 자유롭게 변경하셔서 실제 결과와 비슷한 parameter가 무엇일지 찾아보면 좋을 것 같습니다!
            
            # 저는 지금 타율을 0.4, ops를 0.6 가중치로 두었고, 안타칠 확률은 최소 0.1, 최대 0.3으로 설정했습니다.
            
            prob_hit = 0.1 + 0.25 * (0.3 * avg_norm + 0.7 * ops_norm)
            prob_out = 1 - prob_hit

            # 각 선수별 안타, 2루타, 3루타, 홈런칠 확률 생성합니다. (이것도 0.01, 0.005 숫자를 자유롭게 변경 하시면 됩니다.)
            prob_1B = max(0.01, row['1B'] / total_hits)
            prob_2B = max(0.005, row['2B'] / total_hits)
            prob_3B = max(0.002, row['3B'] / total_hits)
            prob_HR = max(0.005, row['홈런'] / total_hits)

            total_ratio = prob_1B + prob_2B + prob_3B + prob_HR
            prob_1B /= total_ratio
            prob_2B /= total_ratio
            prob_3B /= total_ratio
            prob_HR /= total_ratio
            
            probs = {
                '1B': prob_hit * prob_1B,
                '2B': prob_hit * prob_2B,
                '3B': prob_hit * prob_3B,
                'HR': prob_hit * prob_HR,
                'OUT': prob_out
            }

            total_prob = sum(probs.values())
            if total_prob > 0:
                for k in probs:
                    probs[k] /= total_prob

        player_probs[row['타자명']] = probs

    return player_probs
```

```python
def estimate_at_bats(row):
    sac = row.get('희생번트', 0)
    sf = row.get('희생플라이', 0)
    if row['타율'] > 0:
        est_pa = row['안타'] / row['타율']
    else:
        est_pa = 100 
    est_pa -= sac + sf
    return max(1, est_pa)
```

```python
def simulate_at_bat(player_name, player_probs):
    probs = player_probs[player_name]
    return random.choices(list(probs.keys()), weights=list(probs.values()), k=1)[0]
```

```python
def update_bases(base_state, result):
    runs = 0
    new_bases = [0, 0, 0]

    if result == 'OUT':
        return base_state, runs

    if result == 'HR':
        runs = sum(base_state) + 1
        return [0, 0, 0], runs

    if result == '1B':
        runs += base_state[2]
        new_bases[2] = base_state[1]
        new_bases[1] = base_state[0]
        new_bases[0] = 1
    elif result == '2B':
        runs += base_state[2] + base_state[1]
        new_bases[2] = base_state[0]
        new_bases[1] = 1
    elif result == '3B':
        runs += sum(base_state)
        new_bases[2] = 1

    return new_bases, runs
```

```python
def simulate_half_inning(lineup, player_probs, start_idx):
    outs = 0
    bases = [0, 0, 0]
    score = 0
    batter_idx = start_idx
    batter_stats = {name: {'타수': 0, '안타': 0} for name in lineup}

    while outs < 3:
        batter = lineup[batter_idx % len(lineup)]
        result = simulate_at_bat(batter, player_probs)
        batter_stats[batter]['타수'] += 1
        if result != 'OUT':
            batter_stats[batter]['안타'] += 1
        bases, runs = update_bases(bases, result)
        score += runs
        if result == 'OUT':
            outs += 1
        batter_idx += 1

    return score, batter_idx, batter_stats
```

```python
def simulate_game(away_lineup, away_probs, home_lineup, home_probs):
    away_idx, home_idx = 0, 0
    away_scores, home_scores = [], []
    away_stats_total = {name: {'타수': 0, '안타': 0} for name in away_lineup}
    home_stats_total = {name: {'타수': 0, '안타': 0} for name in home_lineup}

    for inning in range(9):
        away_score, away_idx, away_stats = simulate_half_inning(away_lineup, away_probs, away_idx)
        for name in away_stats:
            away_stats_total[name]['타수'] += away_stats[name]['타수']
            away_stats_total[name]['안타'] += away_stats[name]['안타']
        away_scores.append(away_score)

        # 홈팀이 이기고 있는경우, 9회말은 진행하지 않습니다.
        
        if inning == 8 and sum(home_scores) > sum(away_scores):
            break

        home_score, home_idx, home_stats = simulate_half_inning(home_lineup, home_probs, home_idx)
        for name in home_stats:
            home_stats_total[name]['타수'] += home_stats[name]['타수']
            home_stats_total[name]['안타'] += home_stats[name]['안타']
        home_scores.append(home_score)

    return {
        'away_total': sum(away_scores),
        'home_total': sum(home_scores),
        'away_inning_scores': away_scores,
        'home_inning_scores': home_scores,
        'away_stats': away_stats_total,
        'home_stats': home_stats_total
    }
```

```python
def plot_inning_scores(team_name, scores):
    innings = list(range(1, len(scores)+1))
    plt.plot(innings, scores, marker='o', label=team_name)
    plt.xticks(range(1, 10))
    plt.xlabel("Innings")
    plt.ylabel("Scores")
    plt.title("Scores by Inning")
    plt.legend()
    plt.grid(True)
```

```python
def print_player_stats(team_name, stats_dict):
    print(f"\n📊 {team_name} 타자 성적")
    print("-" * 30)
    for player, stats in stats_dict.items():
        ab = stats['타수']
        hits = stats['안타']
        avg = hits / ab if ab > 0 else 0
        print(f"{player}: {ab}타수 {hits}안타 ({avg:.3f})")
```
